Analyzing an affiliated credit card (special card only available to members of a specific group). Think of it like this- the credit card company is like Citibank and the affiliated group is a company like Indian Oil or Barista. The affiliated group has to pay some fees to credit card company (because the credit card company is providing this service to the group)

```python
#importing neccessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Consider parameters for the above problem:

    Average card balance is Rs 1000.
    Each card has an interest rate of 15%
    Membership fee is Rs 20 per card
    Loss rate is 3%
    Rs 25 operating cost per card
    Rs 10 affiliation fee per card (cost to the group organization itself, so that their members get this awesome card program)
    6.5% cost of funds (that credit card company must pay to the bank where it gets money)
    Bill/statement is issued on 1st of every month
    Credit-free period = 15 days from the bill/statement issue date.

Considering our timeline starts from 1st of January,2019. We calculate the revenue,Net profit/loss by the end of the year,that is, December,2019.

```python
#initializing the parameters on monthly basis

avg_bal=1000 #average card balance
rate=15/100 #interest rate of 15% per annum
mem_fee=20*12 #assuming Membership fee monthly---converting it annually
loss_rate=3/100 #loss rate
opr_cost=25*12 #assuming operating cost per card monthly---converting it annually.
affl_card=10*12 #assuming affiliation cost per card monthly---converting it annually.
cof_r=6.5/100 #Cost of funds rate
credit_free=15 #credit-free period
```

Consider 1000 customers for the above scenario, divide these 1000 customers into following groups(upon your assumption and understanding about the credit lending industry)

>Customer who pay bill within the credit-free period

>Customers paying 30 days after the due-date.

>Customer paying 45 days after the last due date.

>Customer paying 60 days after the last due date (remove these customers for the complete calculation for the next month, basically you are not supposed to provide further credit services to these customer)

Assumptions made on the basis of number of customers, with the aforementioned conditions are stated below:

The customers who pay bill within credit-free period are **'90%'** of the total customers in the database.

The customers paying 30 days after the last due-date are **'5%'** of the total customers in the database.

The customer paying 45 days after the last due date are **'3%'** of the total customers in the database.

The customer paying 60 days after the last due date are **'2%'** of the total customers in the database.

```python
#Let's initialize the parameters
active_cust=1000  #Customers who are present in the database after the removal of the customers who are paying the credit-bill 60 days after the due-date.
cust=1000 #total customers

def cust_update(active_cust):
    #let the customers who pay bill within credit-free period are 90%
    global cust_no_delay
    cust_no_delay=round(active_cust*(90/100))

    #Let the customers paying 30 days after the due-date are 5%
    global cust_30_delay
    cust_30_delay=round(active_cust*(5/100))

    #Let the customer paying 45 days after the last due date are 3%
    global cust_45_delay
    cust_45_delay=round(active_cust*(3/100))

    #Let the customer paying 60 days after the last due date are 2%
    global cust_60_delay
    cust_60_delay=round(active_cust*(2/100))
    return active_cust
```

```python
cust_update(active_cust)
```

Category-wise distribution of customers are done with the help of list **'category_type'** and **'Category_cust'**. 

```python
category_type=[cust_no_delay,cust_30_delay,cust_45_delay,cust_60_delay]
Category_cust=['Paid with no delay','Paid after 30 days','Paid after 45 days','Paid after 60 days']
```

Calculation of interest after 'n' days of payement:
$$
Interest(n) = (n * OutstandingBalance * InterestRatePerYear)/365
$$

**Calculation of Interest**

A user-defined function **'interest_amount(n)'** is made which calculates the Interest for each customer category where n is customer category. 

```python
#Billing cycle starts from 1st of every month
#Calculating monthly interest

def interest_amount(n):
    if n==category_type[0]: #customers paying credit card bill within due date
        interest=0
    elif n==category_type[1]:  #customers paying credit card bill 30 days after last due date
        interest=(((credit_free + 30)*avg_bal*rate)/365)*category_type[1]
    elif n==category_type[2]:   #customers paying credit card bill 45 days after last due date
        interest=(((credit_free + 45)*avg_bal*rate)/365)*category_type[2]
    else:
        interest=(((credit_free + 60)*avg_bal*rate)/365)*category_type[3]  #customers paying credit card bill 60 days after last due date
        #Remove these customers from the credit-service
    return interest 
```

A function **'calculations()'** is described which is responsible for the calculation of the interest, Principal Amount and total Amount. All the parameters are list which are getting appended with every iteration till the 4th iteration. 


```python
#Categorywise calculation of Interest and other parameters:

def calculations():
   global interest
   interest=[] #stores the interest of the active customers
   principal=[] #stores the principal amount of the active customers
   global Amount
   Amount=[] #stores the overall amount(principal + interest) of the active customers
   Amount.clear()
   principal.clear()
   interest.clear()
   for i in range(4):
      Amount.append(category_type[i]*avg_bal + interest_amount(category_type[i]))
      principal.append(category_type[i]*avg_bal)
      interest.append(interest_amount(category_type[i]))
   return Amount
```

Creating an empty dataframe, **'data_new'** with the column names *'Category_cust','Number of customers','Amount','Interest','month'*

This dataframe will be further used to concatenate temporary dataframes.

```python
data_new=pd.DataFrame(columns={'Category_cust','Number of customers','Amount','Interest','month'})
data_new
```

* By the second week of April(12th April,2019), we would have the our list of customers falling in different categories mentioned above. We will revoke credit services of the customers falling in the 4th category(defaulters) from the list of customers and thereby our dataframe would be updated accordingly.

* This is how the below piece of code works for every cycle after April. Hence the 'month' column in the dataframe starts with the number 4, that means the 4th month, i.e, April.

```python
for i in range(9):
    print("Active customers in the dataframe:",active_cust)
    cust_update(active_cust)
    category_type=[cust_no_delay,cust_30_delay,cust_45_delay,cust_60_delay]
    calculations()
    data_temp=pd.DataFrame({'Category_cust': Category_cust,'Number of customers':category_type,'Amount': Amount,'Interest':interest,'month':i+4})
    data_new=pd.concat([data_new,data_temp])
    active_cust=active_cust - cust_60_delay
    cust_update(active_cust)
data_new
```
